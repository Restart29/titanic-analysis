## 📌 Impute missing 'Cabin' values based on shared 'Ticket' values
------------------------------------------------------------------------------
Step 1: Create a mapping from each Ticket number to a known (non-null) Cabin.
- Group the data by 'Ticket'.
- For each group, check if any Cabin value is present.
- If yes, extract the first unique non-null cabin.
- If not, assign NaN (this group doesn't help us impute anything).
ticket_to_cabin = db.groupby('Ticket')['Cabin'].apply(
    lambda x: x.dropna().unique()[0] if x.notna().any() else np.nan
)

##### Step 2: Use the mapping to fill in missing Cabin values
- For each row with missing Cabin, check if their Ticket is in the mapping.
- If so, fill in the corresponding Cabin from the mapping.
- If not (or the ticket had no cabin associated), it stays NaN.
db['Cabin'] = db['Cabin'].fillna(db['Ticket'].map(ticket_to_cabin))


### Cabin Imputation for Likely Servant Passengers

  - Same ticket number and fare
  - 0 SibSp and 0 Parch
  - Different last names
- Only one passenger in each pair had a cabin assigned
- Based on contextual reasoning (possible servant-employer relationship), assigned the same cabin to the unlisted passenger

## 🛠️ Cabin Feature Cleaning & Engineering

- Imputed missing `Cabin` values based on shared `Ticket` numbers, assuming passengers sharing a ticket may belong to the same group or family (e.g., servants).
- Extracted the `Deck` from the `Cabin` string using regex, selecting only the first capital letter to represent the deck level (e.g., 'C123' → 'C').
- Filled missing `Deck` entries with `'Unknown'` to preserve missingness information.
- Created a new binary column `HasCabin`, indicating whether a cabin assignment was present or not (`1` if cabin exists, `0` otherwise).
